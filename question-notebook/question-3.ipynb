# Q3. 手推反向传播

对于标量 $\text{loss} = (\hat{y} - y)^2$，其中 $\hat{y} = wx + b$：

1. 手算 $\frac{\partial \text{loss}}{\partial w}$ 和 $\frac{\partial \text{loss}}{\partial b}$
2. 用 PyTorch autograd 验证，打印 `w.grad` 和 `b.grad`，与手算结果对比

$$x = 2.0,\quad y = 5.0,\quad w = 1.5,\quad b = 0.0$$
$$\text{loss} = (wx + b - y)^2$$

**验收标准**：手算值与 `.grad` 值完全一致。

## 手算推导

**前向传播：**

$$\hat{y} = wx + b = 1.5 \times 2.0 + 0.0 = 3.0$$

$$\text{loss} = (\hat{y} - y)^2 = (3.0 - 5.0)^2 = 4.0$$

**反向传播（链式法则）：**

$$\frac{\partial \text{loss}}{\partial \hat{y}} = 2(\hat{y} - y) = 2(3.0 - 5.0) = -4.0$$

$$\frac{\partial \hat{y}}{\partial w} = x = 2.0 \qquad \frac{\partial \hat{y}}{\partial b} = 1$$

$$\frac{\partial \text{loss}}{\partial w} = \frac{\partial \text{loss}}{\partial \hat{y}} \cdot \frac{\partial \hat{y}}{\partial w} = -4.0 \times 2.0 = \boxed{-8.0}$$

$$\frac{\partial \text{loss}}{\partial b} = \frac{\partial \text{loss}}{\partial \hat{y}} \cdot \frac{\partial \hat{y}}{\partial b} = -4.0 \times 1 = \boxed{-4.0}$$

**直觉：** 梯度为负说明增大 $w$ 或 $b$ 会让 loss 下降（因为现在预测值 3.0 比目标 5.0 小，需要往大的方向调）。

```python
import torch

x = torch.tensor(2.0)
y = torch.tensor(5.0)

w = torch.tensor(1.5, requires_grad=True)
b = torch.tensor(0.0, requires_grad=True)

# Forward
y_hat = w * x + b
loss  = (y_hat - y) ** 2

print(f"y_hat = {y_hat.item()},  loss = {loss.item()}")

# Backward
loss.backward()

print(f"w.grad = {w.grad.item()}  (expected: -8.0)")
print(f"b.grad = {b.grad.item()}  (expected: -4.0)")
print(f"Match: {w.grad.item() == -8.0 and b.grad.item() == -4.0}")
```
