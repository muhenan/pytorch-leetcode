# RELU

```python
import torch
import matplotlib.pyplot as plt
```

## 手写 ReLU

ReLU (Rectified Linear Unit) 定义：

$$\text{ReLU}(x) = \max(0, x)$$

```python
def my_relu(x: torch.Tensor) -> torch.Tensor:
    return (x > 0) * x
```

## 验证：手写 vs torch 内置

```python
x = torch.linspace(-5, 5, 200)

y_mine = my_relu(x)
y_torch = torch.relu(x)

# 数值验证
print("手写 ReLU 与 torch.relu 结果一致：", torch.allclose(y_mine, y_torch))
```

```python
x1 = torch.tensor([-2., -1., 0., 1., 2.])
print("Input: ", x1)
print("Output:", my_relu(x1))
```

## 画图对比

```python
plt.plot(x.numpy(), y_mine.numpy(), color='steelblue', linewidth=2)
plt.axhline(0, color='gray', linewidth=0.8, linestyle='--')
plt.axvline(0, color='gray', linewidth=0.8, linestyle='--')
plt.title('手写 ReLU (my_relu)')
plt.xlabel('x')
plt.ylabel('ReLU(x)')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
plt.plot(x.numpy(), y_torch.numpy(), color='tomato', linewidth=2)
plt.axhline(0, color='gray', linewidth=0.8, linestyle='--')
plt.axvline(0, color='gray', linewidth=0.8, linestyle='--')
plt.title('torch.relu')
plt.xlabel('x')
plt.ylabel('ReLU(x)')
plt.grid(True, alpha=0.3)
plt.show()
```
